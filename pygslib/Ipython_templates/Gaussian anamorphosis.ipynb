```python
#general imports
import pygslib    
import pandas as pd

#make the plots inline
%matplotlib notebook 
```

```python
#get the data in gslib format into a pandas Dataframe
clusterdata= pygslib.gslib.read_gslib_file('../datasets/cluster.dat') 
clusterdata.head()
```

## Interactive anamorphosis modeling

```python
# Fit anamorphosis by changing, zmax, zmin, and extrapolation function
PCI, raw, zana, gauss, raw_var, PCI_var, ax1 = pygslib.nonlinear.anamor(
                         z = clusterdata['Primary'], 
                         w = clusterdata['Declustering Weight'], 
                         zmin = clusterdata['Primary'].min(), 
                         zmax = clusterdata['Primary'].max()+10,
                         zpmin = 0.01, zpmax = 200,
                         ymin=-5, ymax=5,
                         ndisc = 5000,
                         ltail=1, utail=4, ltpar=1, utpar=1.5, K=40, epson=0.01)
```

## Anamorphosis modeling from raw Z,Y pairs

```python
PCI, raw, zana, gauss, raw_var, PCI_var, ax2 = pygslib.nonlinear.anamor_raw(
                         z = clusterdata['Primary'], 
                         w = clusterdata['Declustering Weight'], 
                         K=40)
```

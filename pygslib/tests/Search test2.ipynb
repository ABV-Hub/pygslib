========

Introduction
---------------
K-dimensional trees (KDTREE) is a space-partitioning data structure for organizing points in a k-dimensional space. This is useful to do nearest neighbor searches. A well documented python module for KDTREE is provided in scipy: scipy.spatial.KDTree.

The implementation
-----------------

The Idea is rotate the data and target points to impose isotropy. In other words, we want a serch sphere instead of a search ellipse.

The algorithm is: 
1. Define a pivot point for rotation (example a corner of the grid)
1. Rotate the data and target point with the same rotation definition of the search ellipse and create a temporary coordinate system x', y', z'
1. Calculate the anisotropy in the y'and z' directions and rescale (multiply by aniso_y and aniso_z)
1. Build the kdtree: tree = scipy.spatial.KDTree(data)
1. For each point in target (or a set of targets for parallel computing)
   - find the n nearest neighbors to the target d,i=KDTree.query(target point, k=The number of nearest neighbors to return, eps=tolerance, p=2, distance_upper_bound=inf)
   - or find all points within a radius with d,i=KDTree.query_ball_point(x, r, p=2.0, eps=0)
   - apply refinement (max per bhid, octants)
   

>scipy.spatial.KDTree.query_ball_point results : list or array of lists
If x is a single point, returns a list of the indices of the neighbors of x. If x is an array of points, returns an object array of shape tuple containing lists of neighbors. 

>This allows to find a set of sets of datapoints for multiple target points in a single run. This is handy if using small search ellipses


Interesting! 

scipy.spatial.KDTree has some useful methods methods for example for variography... and EDA.

- count_neighbors(other, r[, p])	Count how many nearby pairs can be formed.	
- query_ball_tree(other, r[, p, eps])	Find all pairs of points whose distance is at most r
- query_pairs(r[, p, eps])	Find all pairs of points within a distance.
- sparse_distance_matrix(other, max_distance)	Compute a sparse distance matrix

```python
#general imports
import matplotlib.pyplot as plt   
import pygslib as gslib   
import pandas as pd
import numpy as np
from scipy import spatial
from matplotlib.patches import Ellipse

#make the plots inline
%matplotlib inline  
```

Getting the data ready for work
---------
If the data is in GSLIB format you can use the function `gslib.read_gslib_file(filename)` to import the data into a Pandas DataFrame. 


```python
#get the data in gslib format into a pandas Dataframe
mydata= pd.DataFrame({'Xlocation':np.linspace(0, 100, 100), 
                      'Ylocation':np.linspace(0, 100, 100), 
                      'Zlocation':np.linspace(0, 100, 100)})
```

```python
#view data in a 2D projection
plt.scatter(mydata['Xlocation'],mydata['Ylocation'], c=mydata['Zlocation'])
l=plt.colorbar()
l.set_label('Elevation')
plt.grid(True)
plt.show()
```

##Search n-close and sircle ( rotation and anisotropy)

```python
# define rotation

tx=50
ty=50
tz=50
anis1 = 0.6
anglez = 45

parameters_data = { 
                    'x'      :  mydata.Xlocation,     # data x coordinates, array('f') with bounds (na), na is number of data points
                    'y'      :  mydata.Ylocation,     # data y coordinates, array('f') with bounds (na)
                    'z'      :  mydata.Zlocation,     # data z coordinates, array('f') with bounds (na)
                    'x0'     :  0,           # new X origin of coordinate , 'f' 
                    'y0'     :  0,           # new Y origin of coordinate , 'f'
                    'z0'     :  0,           # new Z origin of coordinate , 'f'
                    'ang1'   :  anglez,           # Z  Rotation angle, 'f' 
                    'ang2'   :  0.,           # X  Rotation angle, 'f'
                    'ang3'   :  0.,           # Y  Rotation angle, 'f'
                    'anis1'  :  anis1,           # Y cell anisotropy, 'f' 
                    'anis2'  :  1.,           # Z cell anisotropy, 'f' 
                    'invert' :  0}            # 0 do rotation, <> 0 invert rotation, 'i'

parameters_target = { 
                    'x'      :  [tx],     # data x coordinates, array('f') with bounds (na), na is number of data points
                    'y'      :  [ty],     # data y coordinates, array('f') with bounds (na)
                    'z'      :  [tz],     # data z coordinates, array('f') with bounds (na)
                    'x0'     :  0,           # new X origin of coordinate , 'f' 
                    'y0'     :  0,           # new Y origin of coordinate , 'f'
                    'z0'     :  0,           # new Z origin of coordinate , 'f'
                    'ang1'   :  anglez,           # Z  Rotation angle, 'f' 
                    'ang2'   :  0.,           # X  Rotation angle, 'f'
                    'ang3'   :  0.,           # Y  Rotation angle, 'f'
                    'anis1'  :  anis1,           # Y cell anisotropy, 'f' 
                    'anis2'  :  1.,           # Z cell anisotropy, 'f' 
                    'invert' :  0}            # 0 do rotation, <> 0 invert rotation, 'i'



mydata['xr'],mydata['yr'],mydata['zr']  = gslib.rotscale(parameters_data)
txr,tyr,tzr  = gslib.rotscale(parameters_target)

print txr,tyr,tzr

tree = spatial.KDTree(mydata[['xr','yr','zr']])


#select the 200 points close to 25,25 and within 10m 
d,i= tree.query([txr[0],tyr[0],tzr[0]], k=200, distance_upper_bound=10)

#need some cleanup (d= infinite if > 10)
i = i[np.isfinite(d)]
d = d[np.isfinite(d)]


#this is the data selected:
#print mydata.iloc[i]

#clean up...
fig= plt.figure(num=None, figsize=(12, 10), dpi=200, facecolor='w', edgecolor='k')
ax = fig.add_subplot(111)


e1 = Ellipse((tx, ty), width=20.*anis1, height=20.,
             angle=-anglez, linewidth=2, fill=False, zorder=2)

ax.add_patch(e1)

plt.scatter(mydata['Xlocation'].iloc[i],mydata['Ylocation'].iloc[i],  s=d*100, c=d, alpha=0.5, label = 'Selected data')
plt.plot(mydata['Xlocation'],mydata['Ylocation'], '.', color='g',  label = 'Data')
plt.plot(tx,ty, 'r*', label= 'Target point')
l=plt.colorbar()
l.set_label('Distance to target')
plt.legend(loc=4)
plt.grid(True)
plt.show()
```

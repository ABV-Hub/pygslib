```python
#general imports
import matplotlib.pyplot as plt  
import pygslib as pygsl
import numpy as np

#make the plots inline
%matplotlib inline  
```

# The dataset
Before we can do anything we need a dataset, you can import any format supported by numpy, pandas. For GSLIB format we have an especial function...

```python
mydata= pygsl.read_gslib_file('../datasets/cluster.dat') 
```

```python
mydata.head()
```

# Updating parameters

We will change some parameter:
 - `koption` to 1, this is cross validation a
 - `idbg` to 0. 
 - the input data and the output data with the same variables
 - the discretization points `nxdis` to one 

```python
kt3d_Parameters = {
            # Input Data 
            # ----------
            'x' : mydata['Xlocation'],   # rank-1 array('f') with bounds (nd)
            'y' : mydata['Ylocation'],   # (optional) rank-1 array('f') with bounds (nd)
            'vr' : mydata['Primary'],   # (optional) rank-1 array('f') with bounds (nd)
            # Output (Target) 
            # ----------
            'nx' : 2,  # int
            'ny' : 2,  # int
            'nz' : 2, # int
            'xmn' : 0,   # float
            'ymn' : 0,  # float
            'zmn' : 0,  # float
            'xsiz' : 1,  # float
            'ysiz' : 1,  # float
            'zsiz' : 1,  # float
            'nxdis' : 1,  # int
            'nydis' : 1,  # int
            'nzdis' : 1,  # int
            'outx' : mydata['Xlocation'],  # rank-1 array('f') with bounds (nout)
            'outy' : mydata['Ylocation'],  # (optional) rank-1 array('f') with bounds (nout)
            # Search parameters 
            # ----------
            'radius'     : 20,   # float
            'radius1'    : 20,   # float
            'radius2'    : 20,   # float
            'ndmax'      : 8,   # int
            'ndmin'      : 4,   # int
            # Kriging parameters and options 
            # ----------
            'koption'    : 1,   # (optional) int
            'ktype'      : 1,   # (optional) int
            'idbg'       : 0,   # (optional) int
            # Variogram parameters 
            # ----------
            'c0'         : 0.2,   # float
            'it'         : [1],   # rank-1 array('i') with bounds (nst)
            'cc'         : [0.8],   # rank-1 array('f') with bounds (nst)
            'aa'         : [10],   # rank-1 array('f') with bounds (nst)
            'aa1'        : [10],   # rank-1 array('f') with bounds (nst)
            'aa2'        : [10]}   # rank-1 array('f') with bounds (nst)
```

# Running the estimate

```python
#run the estimate 
estimate, debug, summary = pygsl.gslib.kt3d(kt3d_Parameters)
```

```python
# this is the result

import pandas as pd
result= pd.DataFrame(estimate)
result ['Actual'] =  mydata['Primary']
result ['error'] = result['Actual']- result['outest']
result.head()
```

```python
result ['error'].hist()
```

```python
ax = plt.subplot()
plt.plot(result ['Actual'], result ['error'], 'o')
plt.plot([result ['Actual'].min(),result['Actual'].max()],[0,0], 'k-')
ax.set_xscale("log", nonposx='clip')
ax.set_ylabel ('Error')
ax.set_xlabel ('Actual')
ax.set_title ('Actual vs Error')
```

```python
ax = plt.subplot()
plt.plot(result ['Actual'], result ['outest'], 'o')
plt.plot([result ['Actual'].min(),result['Actual'].max()],[result ['Actual'].min(),result['Actual'].max()], 'k-')
ax.set_xscale("log", nonposx='clip')
ax.set_yscale("log", nonposx='clip')
ax.set_ylabel ('Estimate')
ax.set_xlabel ('Actual')
ax.set_title ('Actual vs Estimate')
```
